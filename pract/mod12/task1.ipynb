```python
import sqlite3
import pandas as pd
conn_netflix = sqlite3.connect('netflix.sqlite')
df = pd.read_sql("SELECT * FROM netflix_titles", conn_netflix)
actors = []

df_actor_title = pd.DataFrame(columns = ['show_id', 'actor_id'])
i = 0
for index, row in df.iterrows():
    cast = row['cast'].split(', ')
    show_id = row['show_id']
    for actor in cast:
      if actor not in actors:
        actors.append(actor)
      actor_id = actors.index(actor)
      df_actor_title.loc[i] = [show_id, actor_id]
      i += 1
df_actor_title.head(5)
```

```python
df_actors = pd.DataFrame(actors)
df.pop('cast')
df.head(5)
```

```python
conn = sqlite3.connect('task1.sqlite')
df.to_sql('titles', conn)
df_actors.to_sql('actors', conn)
df_actor_title.to_sql('actor_title', conn)
```

```python
df_test = pd.read_sql('''SELECT
    a1.[0] AS first_actor,
    a2.[0] AS second_actor,
    COUNT(*) AS times_worked_together
FROM
    actor_title at1
JOIN
    actor_title at2 ON at1.show_id = at2.show_id AND at1.actor_id < at2.actor_id
JOIN
    actors a1 ON at1.actor_id = a1.[index]
JOIN
    actors a2 ON at2.actor_id = a2.[index]
GROUP BY
    a1.[0], a2.[0]
ORDER BY
    times_worked_together DESC
LIMIT 1;''', conn)
print(df_test)
```
